## Normalizing Financial Time series to a Base value

```python
close.iloc[0 ,0]
```

To normalize the price, here is a one method call normalizing where we divide each close value by initial value. It means that we invested $1 to particular ticker symbol, how much return we should have received. 

```python
close.BA.div(close.iloc[0, 0])
```

```python
close.BA.div(close.iloc[0, 0]).mul(100)
```

```python
norm = close.div(close.iloc[0]).mul(100)
norm
```

```python
norm.dropna().plot(figsize=(15, 8), fontsize= 13, logy= False)
plt.legend(fontsize = 13)
plt.show()
```

```python
norm.dropna().plot(figsize=(15, 8), fontsize= 13, logy= True)
plt.legend(fontsize = 13)
plt.show()
```

```python
msft = yf.download("MSFT", "2014-10-01", "2023-07-08")
msft
```

```python
msft['Close'].diff()
```

```python
msft["P_diff"] = msft.Close.diff()
```

```python
msft
```

```python
msft["Percent_change"] = msft.P_diff / msft['Close'] * 100
msft
```

```python
msft[['Close', 'Percent_change']]
```

```python
msft.Percent_change.plot(figsize=(15, 8), fontsize= 13)
plt.legend(fontsize = 13)
plt.show()
```

```python
import matplotlib.pyplot as plt
pd.options.display.float_format = '{:.4f}'.format
plt.style.use("seaborn")

msft = yf.download("MSFT", "2013-01-01", "2023-07-10")
```

```python
# This shows how much we could have make for every $1 invested in the stock. 
multiple = (msft.Close[-1] / msft.Close[0])
multiple
```

```python
# The return does not give a full story unless there is time period information. 
start = msft.index[0]
end = msft.index[-1]
years_invested = (end - start).days / 365.25
years_invested  
```

```python
cagr = multiple ** (1 / years_invested) - 1
cagr
```
