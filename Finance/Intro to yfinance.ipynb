```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
ticker = ['AAPL', "BA", "KO", "IBM", "DIS", "MSFT"]
```

```python
stocks = yf.download(ticker, start="2010-01-01")
```

```python
stocks.info()
```

```python
stocks.head()
```

```python
stocks.describe()
```

```python
close_price = stocks.loc[:,"Close"]
```

```python
close_price.plot(figsize=(15, 8), fontsize=13)
plt.legend(fontsize=15)
plt.show()
```

```python
# Normalizing the Apple stock price
close_price.AAPL.div(close_price.iloc[0,0])
```

```python
# Let's normalize all the stocks with pandas
normalize = close_price.div(close_price.iloc[0])
```

```python
normalize
```

```python
normalize.plot(figsize=(25,15), fontsize=20)
plt.show()
```
