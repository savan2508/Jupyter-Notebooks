# Matplotlib Library visualization  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = 2 * x
```

```python
y
```

```python
plt.plot(x, y)
plt.title("X vs Y")
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.ylabel("Y Axis")
plt.xlabel("X Axis");
```

```python
a = np.linspace(0, 10, 1)
b = a ** 4
```

```python
x = np.arange(0, 10)
y = 2 ** x
```

```python
print(f"a= {a} \n b= {b} \n x = {x} \n y = {y}")
```

```python
# Plotting multiple charts on the same figure
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title("Power of 4")

# Small Axes
axes2 = fig.add_axes([0.2, 0.2, 0.25, 0.25])
axes2.plot(x, y)

plt.show()
```

```python
# Plotting the same charts on the same figure with a zoomed-in version 
fig = plt.figure()

# Large axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a, b)

axes1.set_xlim(0, 8)
axes1.set_ylim(0, 8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title("Power of 4")

# Small Axes
axes2 = fig.add_axes([0.2, 0.2, 0.25, 0.25])
axes2.plot(a, b)

axes2.set_xlim(1, 2)
axes2.set_ylim(0, 50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title("Zoomed In")

plt.show()
```

```python
# We can modify the size of the plot and resolution of the image as follow. 
fig = plt.figure(figsize=(10 ,3), dpi = 300)
axes1 = fig.add_axes([0, 0, 1, 1])
axes1.plot(a, b)
fig.savefig('trial_new_fig_1.png', bbox_inches = 'tight')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, dpi=200)

axes[0][0].plot(x, y)
axes[0][1].plot(y, x, ls="--")
axes[1][0].plot(a, b, linewidth=2,ls="-.")
axes[1][1].plot(b, a, color="red")
axes[1][1].set_title('1 1 Title')
axes[1][1].set_xlabel('1 1 X Label')

fig.subplots_adjust(wspace=0.3, hspace=0.4)
```
