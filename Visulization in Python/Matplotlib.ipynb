# Matplotlib Library visualization  

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,10)
x
```

```python
y = 2 * x
```

```python
y
```

```python
plt.plot(x, y)
plt.title("X vs Y")
plt.xlim(0, 6)
plt.ylim(0, 12)
plt.ylabel("Y Axis")
plt.xlabel("X Axis");
```
