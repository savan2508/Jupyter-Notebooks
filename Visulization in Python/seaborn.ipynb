# Seaborn
* When using seaborn, we trade off customization for ease of use.
* However, since seaborn built directly off of Matplotlib, we can actually make plt method calls to directly affect the resulting seaborn plot. 


## Charting with seaborn
* Scatter plots: Scatter plots show the relationship between two continuous features. Continuous features are numeric variables that can take any number of values between any two values. 
    * For example,
        * Age
        * Height
        * Salary
        * Temperature
    * A continuous feature allows for a value to always be between two values. 
    

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data/dm_office_sales.csv")
df
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df)
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue='level of education')
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df, hue="salary")
```

```python
plt.figure(figsize=(12, 4), dpi=200)
sns.scatterplot(x='salary', y='sales', data=df)
```
