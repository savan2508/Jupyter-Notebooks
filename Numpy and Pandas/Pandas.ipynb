```python
# Create a new column for the tip percentage
df["tip_percentage"] = np.round(100 * df["tip"] / df["total_bill"], 2)
df.head()
```

```python
# To change the index to column which have distinct values
df.set_index("Payment ID")
```

```python
df.iloc[0]
```

```python
# Get a last four of the credit card number
def last_four(num):
    return str(num)[-4:]


df["last_four"] = df["CC Number"].apply(last_four)
df.head()
```

```python
df["total_bill"].mean()
```

```python
def yelp(price):
    if price < 10:
        return "$"
    elif price >= 10 and price < 20:
        return "$$"
    else:
        return "$$$"


df["Price_rating"] = df["total_bill"].apply(yelp)
df
```

```python
df["total_bill"].apply(lambda bill: bill * 2)
```

```python
def quality(total_bill, tip):
    if tip / total_bill > 0.25:
        return "Generous"
    else:
        return "Other"


df["Tip_Quality"] = df[["total_bill", "tip"]].apply(
    lambda df: quality(df["total_bill"], df["tip"]), axis=1
)
```

```python
df.head()
```

```python
df["Quality"] = np.vectorize(quality)(df["total_bill"], df["tip"])
```
