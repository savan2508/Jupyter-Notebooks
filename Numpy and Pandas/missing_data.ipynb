# Missing Data


There are various methods are recommended when dealing with any missing data in dataset. Options for the missing data:

- Keep it
- Remove it
- Replace it

(There are no recommended methods to utilize in case of missing data. The user pick best available option based on the experience.)


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("files/movie_scores.csv")
df
```

```python
df[df["pre_movie_score"].notnull()]
```

```python
df[(df["pre_movie_score"].isnull()) & (df["first_name"].notnull())]
```

## Drop the values


```python
# Drop all the rows which have the missing values
df.dropna()
```

```python
# Provide argument to keep the column if at least one of the column has any data.
df.dropna(thresh=1)
```

```python
# Drop empty cells from the last_name column
df.dropna(subset=["last_name"])
```
